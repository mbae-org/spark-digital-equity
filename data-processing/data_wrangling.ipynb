```python
import pandas as pd
import numpy as np
```

```python
course_enrollment = pd.read_csv('initial_data/CS Course Enrollment 2018.csv')
```

```python
course_enrollment = course_enrollment.drop(columns=["# STUDENTS ENROLLED PRIOR TO SECONDARY LEVEL", "# STUDENTS ENROLLED AT SECONDARY LEVEL"])
```

```python
demographic = pd.read_csv('initial_data/AP CS Gender Ethnicity 2018.csv')
```

```python
demographic
```

```python
demographic.dtypes
```

```python
demographic = demographic.drop(columns=['DIST_NAME', 'SCH_NAME'])
```

```python
demographic
```

```python
demographic["AA"] = pd.to_numeric(demographic["AA"], errors='coerce')
demographic["AS"] = pd.to_numeric(demographic["AS"], errors='coerce')
demographic["HI"] = pd.to_numeric(demographic["HI"], errors='coerce')
demographic["MR"] = pd.to_numeric(demographic["MR"], errors='coerce')
demographic["NA"] = pd.to_numeric(demographic["NA"], errors='coerce')
demographic["NH_PI"] = pd.to_numeric(demographic["NH_PI"], errors='coerce')
demographic["WH"] = pd.to_numeric(demographic["WH"], errors='coerce')
demographic["EcoDis"] = pd.to_numeric(demographic["EcoDis"], errors='coerce')
demographic["SWD"] = pd.to_numeric(demographic["SWD"], errors='coerce')
demographic["ELL"] = pd.to_numeric(demographic["ELL"], errors='coerce')
demographic["FEMALE"] = pd.to_numeric(demographic["FEMALE"], errors='coerce')
demographic["MALE"] = pd.to_numeric(demographic["MALE"], errors='coerce')
demographic["NONBINARY"] = pd.to_numeric(demographic["NONBINARY"], errors='coerce')
```

```python
demographic["AA"] = demographic["AA"].astype("Int64")
demographic["AS"] = demographic["AS"].astype("Int64")
demographic["HI"] = demographic["HI"].astype("Int64")
demographic["MR"] = demographic["MR"].astype("Int64")
demographic["NA"] = demographic["NA"].astype("Int64")
demographic["NH_PI"] = demographic["NH_PI"].astype("Int64")
demographic["WH"] = demographic["WH"].astype("Int64")
demographic["EcoDis"] = demographic["EcoDis"].astype("Int64")
demographic["SWD"] = demographic["SWD"].astype("Int64")
demographic["ELL"] = demographic["ELL"].astype("Int64")
demographic["FEMALE"] = demographic["FEMALE"].astype("Int64")
demographic["MALE"] = demographic["MALE"].astype("Int64")
demographic["NONBINARY"] = demographic["NONBINARY"].astype("Int64")
```

```python
demographic.dtypes
```

```python
scores = pd.read_csv('initial_data/ap_performance-2018.csv')
```

```python
scores
```

```python
scores = scores.drop(columns=['School Name', '% Score 1-2', '% Score 3-5'])
```

```python
scores.dtypes
```

```python
all_merged = course_enrollment.merge(demographic, how='left', on='School Code').merge(scores, how='left', on='School Code')
```

```python
all_merged
```

```python
all_merged.dtypes
```

```python
all_merged = all_merged.astype({"STUDENTS_ENROLLED": "Int64", "ORG_CODE": "Int64"})
```

```python
all_merged["Tests Taken"] = all_merged["Tests Taken"].astype("Int64")
all_merged["Score=1"] = all_merged["Score=1"].astype("Int64")
all_merged["Score=2"] = all_merged["Score=2"].astype("Int64")
all_merged["Score=3"] = all_merged["Score=3"].astype("Int64")
all_merged["Score=4"] = all_merged["Score=4"].astype("Int64")
all_merged["Score=5"] = all_merged["Score=5"].astype("Int64")
```

```python
all_merged.dtypes
```

```python
all_merged['SY'] = 2017
```

```python
all_merged
```

```python
all_merged
```

```python
all_merged.to_csv("all_merged.csv", index=False)
```

```python
all_merged_json = all_merged.to_json(orient='records')
```
